```python
from test_accuracy.test_nntool_pesq_stoi import build_nntool_graph
import numpy as np
from nntool.api.utils import model_settings, qsnr
```

```python
model_path = "model/denoiser_GRU.onnx"
astats_file = "model/denoiser_GRU_stats.pickle"
```

```python
G = build_nntool_graph(model_path, astats_file, states_as_inout=False)
```

```python
G.draw()
```

```python
input_data = np.ones(257)
res = G.execute_on_target(
    directory="/tmp/test_denoiser",
    input_tensors=[input_data],
    check_on_target=True,
    output_tensors=4,
    print_output=True,
    at_loglevel=2,
    settings=model_settings(
        l1_size=128000,
        l2_size=1300000,
        graph_const_exec_from_flash=True,
        graph_group_weights=True,
        graph_l1_promotion=0
    ),
    tolerance=0.02
)
```

```python
qout = G.execute([input_data], quantize=True, dequantize=False)
```

```python
G.qsnrs(qout, res.output_tensors)
```
